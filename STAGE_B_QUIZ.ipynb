```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
energy_data = pd.read_csv(r'C:\Users\AMY BANIGO\Downloads\energydata_complete2.csv')
df = pd.DataFrame(energy_data)
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

**Dropping [Date & Lights] columns**

```python
df.drop(columns=['lights', 'date'], axis=1, inplace=True)
df
```

**Normalizing dataset and setting target variable as 'Appliances'**

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
normalised_df = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)
normalised_df.head()
```

```python
features_df = normalised_df.drop(columns=['Appliances'])
features_df.head()
```

```python
target = normalised_df['Appliances']
target.head()
```

**Splitting data into 70-30 train-test set**

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(features_df, target, 
                                                    test_size=0.3, random_state=42)
```

**Running Multiple Linear Regression**

```python
from sklearn.linear_model import LinearRegression
linear_model = LinearRegression()

#fitting the model to the training dataset
linear_model.fit(x_train, y_train)
```

```python
#obtaining predictions
predicted_values = linear_model.predict(x_test)
predicted_values
```

```python
y_test.head()
```

**Question 13-17**

```python
#MAE
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test, predicted_values)
round(mae, 2) 
```

```python
#RSS
rss = np.sum(np.square(y_test - predicted_values))
round(rss, 2)
```

```python
#RMSE
from sklearn.metrics import  mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test, predicted_values))
round(rmse, 3)
```

```python
#R2_SCORE
from sklearn.metrics import r2_score
r2_score = r2_score(y_test, predicted_values)
round(r2_score, 2)	
```

```python
#comparing the effects of regularisation using Penalizationn Methods
def get_weights_df(model, feat, col_name):
  #this function returns the weight of every feature
  weights = pd.Series(model.coef_, feat.columns).sort_values()
  weights_df = pd.DataFrame(weights).reset_index()
  weights_df.columns = ['Features', col_name]
  weights_df[col_name].round(3)
  return weights_df
```
